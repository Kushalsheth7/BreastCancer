```python
data.drop(['id'],inplace=True,axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
data['diagnosis'] = label_encoder.fit_transform(data['diagnosis'])  # M=1, B=0
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = data.drop('diagnosis', axis=1)
y = data['diagnosis']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 42, test_size = 0.2)
```

```python
x_test
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
x_test
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
kernel = ['linear', 'poly']
for i in kernel:
    print('kernel is '+ i)
    svc_model = SVC(kernel=i)
    svc_model.fit(x_train, y_train)
    
    svc_train_predict = svc_model.predict(x_train)
    svc_train_report = classification_report(y_train, svc_train_predict)
    print(svc_train_report)

    svc_test_predict = svc_model.predict(x_test)
    svc_test_report = classification_report(y_test, svc_test_predict)
    print(svc_test_report)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'poly']
}

grid_search = GridSearchCV(SVC(), param_grid, verbose=3, cv=5)
grid_search.fit(x_train, y_train)

print("Best parameters found: ", grid_search.best_params_)
```

```python
import joblib
joblib.dump(svc_model, 'svc_model.pkl')
```

```python
train_accuracy = accuracy_score(y_train, svc_train_predict)
test_accuracy = accuracy_score(y_test, svc_test_predict)

print(f"Train Accuracy: {train_accuracy * 100:.2f}%")
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```
